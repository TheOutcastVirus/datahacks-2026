```python
# Cell 4b — convert checkpoint to .ply
import torch
from gsplat.utils import save_ply

ckpt = torch.load(
    "/content/results/ckpts/ckpt_29999_rank0.pt",
    map_location="cpu"
)

# Extract splat params from checkpoint
splats = ckpt["splats"]

# save_ply takes the splats dictionary and a full file path (despite the parameter name 'dir')
save_ply(splats=splats, dir="/content/results/output.ply")
print("Saved ply file to /content/results/output.ply")
```
